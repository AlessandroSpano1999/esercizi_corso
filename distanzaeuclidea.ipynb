### Distanza Euclidea

In matematica, la distanza euclidea √® una distanza tra due punti, in particolare √® una misura della lunghezza del segmento avente per estremi i due punti.

```python
import math
```

$${P}_1({x}_1,{y}_1)\ {P}_2({x}_2,{y}_2)$$

$$d({P}_1,{P}_2) := \sqrt{({x}_2 - {x}_1)^2 + ({y}_2 - {y}_1)^2 }$$

```python
# per lo studente da fare durante in bootcamp : implementa la funzione per il calcolo della distanza euclidea
   # 1. stampa le due coordinate x1 e x2
    # 2. stampa le due coordinate y1 e y2
    # 3. calcola dx e dy (differenza tra le ascisse e le ordinate)
    # 4. calcola la distanza quadrata
    # 5. calcola la radice quadrata e inseriscila in una variabile risultato
    # 6. stampa il risultato
    # 7. ritorna il risultato
    

import math

def DistanzaEuclideaTraDuePunti(x1, y1, x2, y2):
    print("I nostri punti hanno ascissa x1={} e x2={}".format(x1,x2))#stampo i valori delle ascisse x1 e x2
    print("I nostri punti hanno ordinata y1={} e y2={}".format(y1,y2))#stampo i valori delle ordinate y1 e y2
    dx=x2-x1 #differenza ascisse
    dy=y2-y1 #differenza ordinate
    dxy=pow(dx,2)+pow (dy,2) #calcolo la distanza quadrata,se voglio un numero intero posso mettere dxy=int()
    d=math.sqrt (dxy) #calcolo la radice quadrata e la metto in una variabile
    print("La distanza tra di due punti √® ",d)#stampo la variabile
    return d #ritorno la variabile
    

    
    
DistanzaEuclideaTraDuePunti(3,-2,4,-5)

```
