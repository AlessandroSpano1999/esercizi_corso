```python
netflix_df = pd.read_csv('netflix_data.csv')
netflix_df[0:5]
```

```python
netflix_df.head()
```

```python
netflix_df.tail()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))
sns.countplot(x='type', data=netflix_df)
plt.title("Distribuzione del tipo")
plt.show
```

```python
plt.figure(figsize=(10,8))
netflix_df['country'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 paesi produttori di film')
plt.xlabel('Country')
plt.ylabel('Numero film/Tv-show')
plt.show()
```

```python
plt.figure(figsize=(10,8))
netflix_df[netflix_df['type'] == "Movie"]['duration'].plot(kind='hist', rwidth=0.9,bins=30,color='blue')
plt.title('Distribuzione Durata')
plt.xlabel('Durata (min)')
plt.ylabel('Numero di Film')
plt.show()
```

```python
# Fai un dubset del dataframe per il tipo ''Movie'
netflix_df_movies_only = netflix_df[netflix_df['type'] == 'Movie']

# Seleziona solo la colonna di interesse
netflix_movies_col_subset = netflix_df_movies_only[['title', 'country', 'genre', 'release_year', 'duration']]

# Stampa le prime cinque righe del dataset
netflix_movies_col_subset.head()
```

```python
netflix_movies_col_subset = netflix_df_movies_only[netflix_df_movies_only['duration']<60]
netflix_movies_col_subset[:20]
```

```python
# Definisci una lista vuota
colors = []

# Scorri le righe di netflix_movies_col_subset
for lab, row in netflix_movies_col_subset.iterrows():
    if row['genre'] == "Children":
        colors.append("red")
    elif row['genre'] == "Documentaries":
        colors.append("blue")
    elif row['genre'] == "Stand-Up":
        colors.append("green")
    else:
        colors.append("black")

# Ispeziona i primi 10 valori nell'elenco     
colors[0:10]
```
