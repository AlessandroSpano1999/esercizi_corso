Definire una classe **Point** con le seguenti caratteristiche:
    
- Due attributi: x e y, le coordinate del punto sul piano;

- Un costruttore che accetta due parametri, x e y, che inizializzano gli attributi corrispondenti.
    Questi parametri devono avere un valore predefinito di 0.0.
    
- Un metodo *distanza_to_origin()* che restituisce la distanza del punto dall'origine. 
    La formula è $$\sqrt{x^2 + y^2}$$

- Un metodo *reflect()*, che riflette il punto rispetto all'asse x o asse y, e che avrà le seguenti caratteristiche:


    - accetta un parametro *axis* 
    
    - se axis="x", imposta l'attributo y (non è un refuso!) sul valore negativo dell'attributo y
    
    - se axis="y", imposta l'attributo × sul valore negativo dell'attributo x
    
    - per qualsiasi altro valore di asse, stampa un messaggio di errore

```python
    
    import math
    class Point:
        def __init__(self,x=0,y=0):
            self.x=x
            self.y=y
        def distance_to_origin(self):
            d=math.sqrt((self.x)**2 + (self.y)**2)
            return d
        def reflect(self,axis):
                if axis=='x':
                    self.y*=-1
                elif axis=='y':
                    self.x*=-1
                else:
                    print("Axis non può avere valore diverso da x o y")
    pt = Point(x=3.0,y=7) 
    pt.reflect("x")
    print((pt.x, pt.y)) 
    pt.y = 4.0
    print(pt.distance_to_origin())        
            
        
```

```python
import numpy as np                 # v 1.19.2
import matplotlib.pyplot as plt    # v 3.3.2

# Enter x and y coordinates of points and colors
xs = [pt.x]
ys = [pt.y]
colors = ['r']

# Select length of axes and the space between tick labels
xmin, xmax, ymin, ymax = -5, 5, -5, 5
ticks_frequency = 1

# Plot points
fig, ax = plt.subplots(figsize=(10, 10))
ax.scatter(xs, ys, c=colors)

# Draw lines connecting points to axes
for x, y, c in zip(xs, ys, colors):
    ax.plot([x, x], [0, y], c=c, ls='--', lw=1.5, alpha=0.5)
    ax.plot([0, x], [y, y], c=c, ls='--', lw=1.5, alpha=0.5)

# Set identical scales for both axes
ax.set(xlim=(xmin-1, xmax+1), ylim=(ymin-1, ymax+1), aspect='equal')

# Set bottom and left spines as x and y axes of coordinate system
ax.spines['bottom'].set_position('zero')
ax.spines['left'].set_position('zero')

# elimina i bordi
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Create 'x' and 'y' labels placed at the end of the axes
ax.set_xlabel('x', size=14, labelpad=-24, x=1.03)
ax.set_ylabel('y', size=14, labelpad=-21, y=1.02, rotation=0)

# Create custom major ticks to determine position of tick labels
x_ticks = np.arange(xmin, xmax+1, ticks_frequency)
y_ticks = np.arange(ymin, ymax+1, ticks_frequency)
ax.set_xticks(x_ticks[x_ticks != 0])
ax.set_yticks(y_ticks[y_ticks != 0])
```
