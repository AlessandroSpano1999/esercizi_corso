```python
sql_query3=(
'with salaries as \
(select avg(salary) as med_salary,DEPARTMENT_ID as department \
from employees e \
group by e.DEPARTMENT_ID) \
select EMPLOYEE_ID,SALARY,DEPARTMENT_ID \
from EMPLOYEES e \
join salaries on e.DEPARTMENT_ID=salaries.department \
where e.salary>med_salary \
order by e.DEPARTMENT_ID desc \
select avg(salary) as med_salary,DEPARTMENT_ID as department \
from employees e \
group by e.DEPARTMENT_ID \
order by e.DEPARTMENT_ID desc'
)
df=pd.read_sql(sql_query3,conn)
#Display the DataFrame
df.head()
```
